```python
x_org = torch.linspace(0, 2 * np.pi, 100).requires_grad_()
y_org = torch.sin(x_org)

x = [x_org[i] for i in range(len(x_org))]
y = [y_org[i] for i in range(len(y_org))]

print(y[0])

x = [torch.Tensor([i]).requires_grad_() for i in np.linspace(0, 2 * np.pi, 100)]
y = [torch.sin(item) for item in x]

print(y[0])

x_1grad = grad(y, x, create_graph=True)
x_2grad = grad(x_1grad, x)

x_for_plot = [item.item() for item in x]
y_for_plot = [item.item() for item in y]
dydx_for_plot = [item.item() for item in x_1grad]
dydx2_for_plot = [item.item() for item in x_2grad]
plt.plot(x_for_plot, y_for_plot)
plt.plot(x_for_plot, dydx_for_plot)
plt.plot(x_for_plot, dydx2_for_plot)
plt.plot()
```
